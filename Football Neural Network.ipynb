```python
import pandas as pd
import numpy as np
from tqdm import tqdm

```

```python
database = pd.read_csv('DataBase.csv')
oddsX = pd.read_csv('Odds.csv')
```

```python
df = database.drop(columns = 'Unnamed: 0')
oddsZ = oddsX.drop(columns = 'Unnamed: 0')

odds=pd.DataFrame()

odds['H%'] = (1/(oddsZ['OddsH']))/(1/oddsZ['OddsH']+1/oddsZ['OddsD']+1/oddsZ['OddsA'])
odds['D%'] = (1/(oddsZ['OddsD']))/(1/oddsZ['OddsH']+1/oddsZ['OddsD']+1/oddsZ['OddsA'])
odds['A%'] = (1/(oddsZ['OddsA']))/(1/oddsZ['OddsH']+1/oddsZ['OddsD']+1/oddsZ['OddsA'])

display(df)
```

```python
import tensorflow as tf
from tensorflow.keras import layers

print(tf.VERSION)
print(tf.keras.__version__)
```

```python
model = tf.keras.Sequential([

layers.Dense(64, activation='relu'),
layers.Dense(64, activation='relu'),


layers.Dense(3, activation='softmax')])

model.compile(optimizer=tf.train.AdamOptimizer(0.005),
              loss='mse',
              metrics=['binary_accuracy'])
```

```python
label = ['H','D','A']

test_amount = 650

f = df[label]
data = df.drop('H',1)
data = data.drop('D',1)
data = data.drop('A',1)
labels = database[label]

x_train = data[test_amount:]
x_test = data[:test_amount]


y_train = f[test_amount:]
y_test = f[:test_amount]
display(y_train.info)
odds_test = odds[:test_amount]
```

```python
x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=32,
          validation_data=(x_test, y_test))
```

```python
l = model.predict(x_test)
display(l)
predictions = pd.DataFrame(l)

predictions['H'] = f['H']
predictions['D'] = f['D']
predictions['A'] = f['A']
predictions['OddsH'] = odds_test['H%']
predictions['OddsD'] = odds_test['D%']
predictions['OddsA'] = odds_test['A%']
```

```python
predictions.to_csv("new folder/" +'poisson.csv')
```
